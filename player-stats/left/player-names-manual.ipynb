```r
library(tidyverse)

p101 <- read_csv("./p101.csv", show_col_types = FALSE) %>% mutate(
    surname = str_to_title(surname),
    forename = str_to_title(forename),
    player_name = paste(forename, surname, sep = " ")    
)

players_1 <- p101 %>%
    mutate(
        dob = lubridate::dmy(date_of_birth),
        year = case_when(
            !is.na(notes) ~ as.numeric(str_split_i(notes, " ", 2)),
            .default = lubridate::year(dob)
        ),
        quarter = case_when(
            str_detect(notes, "Q") ~ as.numeric(str_extract(notes, "([1-4])")),
            .default = lubridate::quarter(dob)
        ),
        estimated_year = case_when(
            .default = 0,
            str_detect(str_to_lower(notes), "c") ~ 1)
    ) %>%
    select(
        surname,
        forename,
        date_of_birth,
        dob,
        quarter,
        year,
        estimated_year
    ) %>%
    mutate(
        page = 101
    )
```

```r
df <-
    bind_rows(players_1, players_2, players_3, players_4, players_5, players_6, players_7, players_8) %>%
    mutate(
        surname = str_replace(surname, "'(\\w)", function(x) str_to_title(x)),
        surname = str_replace_all(surname, "^(Mc|Mac|A'|D'|O')(.)", "\\1 \\2"),
        surname = str_to_title(surname),
        surname = ifelse(surname != "St John", str_replace(surname, " ", ""), surname),
        year = case_when(
            dob >= dmy("01-01-1990") ~ year - 100,
            surname == "Hilton" & forename == "Harold" ~ year - 200,
            .default = year
        ),
        dob = case_when(
            dob >= dmy("01-01-1990") ~ dob - years(100),
            TRUE ~ dob
        ),
        period = str_extract(forename, "\\d+"),
        forename = str_remove(forename, "\\d+"),
        period = case_when(
            .default = 1,
            !is.na(period) ~ as.numeric(period)
        )
     ) %>%
     rename(
        dob_yr = year,
        dob_qtr = quarter,
        dob_yr_is_est = estimated_year
     ) %>%
    write_csv("./players.csv")
```
