# `scorers-long.csv`

```r
library(tidyverse)

df <- read_csv("../output/scorers-long.csv", show_col_types = FALSE)
results_df <- read_csv("../output/gls_with_games.csv", show_col_types = FALSE)
```

#### List players that have scored 20+ in a season

```r
df %>%
    group_by(season, player_name) %>%
    summarise(
        goals = sum(goals_scored),
        .groups = "drop"
    ) %>%
    arrange(
        desc(goals),
        season,
        player_name
    ) %>%
    filter(
        goals >= 20
    )
```
